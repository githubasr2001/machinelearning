```python
#Training our first machine learning model 
d["Target"]=d.shift(-1)["TMAX"]
d
```

```python
d1=d.iloc[:-1,:].copy()
d1
```

```python
from sklearn.linear_model import Ridge

reg=Ridge(alpha=.1)
```

```python
predictors=['PRCP','SNOW','TMAX','TMIN']
```

```python
train=d1.loc[:"2021-12-31"]
```

```python
test=d1.loc["2022-01-01": ]
```

```python
reg.fit(train[predictors],train["Target"])
```

```python
predictions=reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test["Target"],predictions)
```

```python
#Evaluating our model 
combined=pd.concat([test["Target"],pd.Series(predictions,index=test.index)],axis=1)
combined.columns=["actual","predictions"]
combined
```

```python
combined.plot()
```

```python
reg.coef_
```

```python
#Creating a function to make predictions

def create_predictions(predictors,d1,reg):
    train=d1.loc[:"2021-12-31"]
    test=d1.loc["2022-01-01": ]
    reg.fit(train[predictors],train["Target"])
    predictions=reg.predict(test[predictors])
    error=mean_absolute_error(test["Target"],predictions)
    combined=pd.concat([test["Target"],pd.Series(predictions,index=test.index)],axis=1)
    combined.columns=["actual","predictions"]
    return error,combined
```

```python
#Adding in rolling means 
d1["month_max"]=d1["TMAX"].rolling(5).mean()
d1
```

```python
d1["month_day_max"]=d1["month_max"]/d1["TMAX"]
```

```python
d1["max_min"]=d1["TMAX"]/d1["TMIN"]
```

```python
predictors=['PRCP','SNOW','TMAX','TMIN','month_max','month_day_max','max_min']
```

```python
dp=d1.iloc[5: ,:].copy()
dp
```

```python
error,combined=create_predictions(predictors,dp,reg)
```

```python
error
```

```python
combined.plot()
```

```python
dp["monthly_avg"]=dp["TMAX"].groupby(dp.index.month).apply(lambda x:x.expanding(1).mean())
dp
```

```python
dp["day_of_year_avg"]=dp["TMAX"].groupby(dp.index.day_of_year).apply(lambda x:x.expanding(1).mean())
dp
```

```python
predictors=['PRCP','SNOW','TMAX','TMIN','month_max','month_day_max','max_min','day_of_year_avg']
```

```python
error,combined=create_predictions(predictors,dp,reg)
```
