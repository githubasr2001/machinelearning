```python
chart=sns.barplot(df['movie'],df['gross_world'],palette='Set2')
chart.set_xticklabels(chart.get_xticklabels(),rotation=90)
plt.show()
```

```python
chart1=sns.barplot(df['movie'],df['gross_us'],palette='rocket')
chart1.set_xticklabels(chart1.get_xticklabels(),rotation=90)
plt.show()
```

```python
chart2=sns.barplot(df['movie'],df['opening_weekend_us'],palette='Set1')
chart2.set_xticklabels(chart2.get_xticklabels(),rotation=90)
plt.show()
```

```python
#Movie and WorldWide gross 

movie_ww=df.groupby(['movie','gross_world']).size().reset_index(name='Count')
```

```python
c=sns.barplot(data=movie_ww,x='movie',y='Count',palette='Set2')
c.set_xticklabels(c.get_xticklabels(),rotation=90)
plt.show()
```
